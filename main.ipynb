```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
df=pd.read_csv('pokemon.csv')
df
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df
```

```python
df.info()
```

```python

cols=df[['type1','attack','defense','speed']]
sb.pairplot(cols,hue='type1')
```

```python
cols1=df[['type2','attack','defense','speed']]
sb.pairplot(cols1,hue='type2')
```

```python
#1 what the correlation between speed, attack and defense

plt.title('correlation between speed, attack and defense')
sb.heatmap(df[['speed','attack','defense']].corr(),annot=True)
```

```python
#2 does weight affect the speed

cor=df['weight_kg'].corr(df['speed'])
cor

#the correlation b/w them is very weak and if its affect it affects in a -ve way.
```

```python
#3.average weight of each type 1
temp=df.groupby('type1')['weight_kg'].mean().sort_values(ascending=False)
temp
```

```python
#4. How many pokemons are in each generation
temp=df.groupby('generation')['generation'].count()
temp
```
